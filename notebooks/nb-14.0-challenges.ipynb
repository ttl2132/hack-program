### Challenge 1: 
Execute the code cell below to see an example of how it works. 
Use markdown in the cell after the code-block to describe the function `random_words_api`. Try to be descriptive about what each step of code in this function does, and why it works. 

```python
def random_words_api(nwords=10):
    "no docstring"
    URL = "https://random-word-api.herokuapp.com/word"
    response = requests.get(url=URL, params={"number": nwords})
    return response.json()

# demonstration
random_words_api(5)
```

**Description:**

The first step of this code stores the url for the REST API from a heroku app that returns a random word in URL.
The second step of this code makes the REST call, adding a param nwords that changes the number of random words that are returned. By default, the nwords value is set to 10.
The last step of random_words_api returns the result in JSON format, and since the words are in a list, this is a valid JSON format.

This function works, because the url https://random-word-api.herokuapp.com/word?number=10, replacing 10 with any integer, is a valid REST call. The ?number=10 corresponds to the param given in the second step.

### Challenge 2: 
Use the `random_words_api` function to get 50 random words and store the result as a variable. Write a function that takes the list of words as input and returns a dictionary with the longest word as the key and the length of the longest word as a value. If there is a tie in the length of words then have it return additional words as keys with their lengths as values.

```python
words = random_words_api(50)

def long_word_dict(words):
    """Takes a list of words and returns a dictionary in the form {word: word_length} for the longest words."""
    lengths = [len(x) for x in words]
    max_len = max(lengths)
    word_dict = {}
    for i in range(len(lengths)):
        if lengths[i] == max_len:
            word_dict[words[i]] = lengths[i]
    return word_dict

long_word_dict(words)
```

### Challenge 3: 
Write a function to take the list of words as input and trim all words to be at most 5 characters in length, and return as a list.

```python
def trim_to_five(words):
    """Takes a list of words and trims the words to 5 letters max, then returns the modified list."""
    for i in range(len(words)):
        if len(words[i]) > 5:
            words[i] = words[i][:5]
    return words
```

### Challenge 4: 
Write a function to take a list of words as input and to count the occurrence of all letters in every word and return as a dictionary mapping letters to integers., e.g., {'a': 10, 'b': 3, 'c': 5, ...}.  

```python
def overall_occurrence(words):
    """Takes a list of words and returns a dict as {letter: overall number of letters in the entire list}."""
    letter_dict = {}
    for word in words:
        for char in word:
            if char in letter_dict:
                letter_dict[char]+=1
            else:
                letter_dict[char] = 1
    return letter_dict
```

### Challenge 5:
Use [toyplot](https://toyplot.readthedocs.io/en/stable/tutorial.html) to create a barplot of the occurrences of each letter in your dictionary from the previous challenge. This will represent a histogram of the letters. Play with the size and color of the figure to try to make it look nice.

```python
#data prep and sort
words = random_words_api(50)
words_dict = overall_occurrence(words)
keys = [str(c) for c in words_dict.keys()]
keys.sort()
vals = [words_dict[key] for key in keys]

#bar plot creation
canvas = toyplot.Canvas(width=800, height=400)
axes = canvas.cartesian(label="Letter Occurences", xlabel="Letters", ylabel="Frequency")
mark = axes.bars(vals, style={"fill": "purple"})
axes.x.ticks.locator = toyplot.locator.Explicit(labels=keys)
```

### Challenge 8: 
Use pandas to load a CSV file from https://eaton-lab.org/data/iris-data-dirty.csv and save as a dataframe. Add custom names to the columns in the dataframe, based on the type of values in them (e.g., numeric versus strings). You can come up with any column names you want for these.

```python
# NOTE: It seems like the first row shouldn't actually be the column names, so I appended it to the end of the df.
file_path = "https://eaton-lab.org/data/iris-data-dirty.csv"
df = pd.read_csv(file_path)
first_row = pd.Series(df.columns)
col_types = df.dtypes
col_names = [str(dt) for dt in df.dtypes]
col_count = {}

# used to distinguish between columns with same dtype
for i in range(len(col_names)):
    name = col_names[i]
    if name in col_count:
        col_count[name] += 1
    else:
        col_count[name] = 1

    # dtype of strings are object so this needs to be changed
    if name == "object":
        col_names[i] = f"string_{col_count[name]}"
    else:
        col_names[i] = f"{name}_{col_count[name]}"

df.columns=col_names
# appending the first row to the end
row = {x:y for (x,y) in zip(col_names, first_row)}
dtype_dict = {x:y for (x,y) in zip(col_names, col_types)}
df = df.append(row, ignore_index=True).astype(dtype_dict)
df
```
